```python
import os
from rdkit import Chem
from rdkit import RDConfig
from rdkit.Chem import Draw
from rdkit.Chem import rdMMPA
path =  os.path.join(RDConfig.RDDocsDir, "Book/data/cdk2.sdf")
mols = [mol for mol in Chem.SDMolSupplier(path) if mol != None]
```

```python
mols[8]
```

```python
# mmps = rdMMPA.FragmentMol(mols[10], maxCuts=1, resultsAsMols=False)
mmps = rdMMPA.FragmentMol(mols[8], maxCuts=2)
```

```python
def get_mmp_list(mmps):
    mmps_list=[]
    for i in range(len(mmps)):
        mmps_list.append(mmps[i][1])
    return mmps_list
```

```python
mmps_list = get_mmp_list(mmps)
```

```python
Draw.MolsToGridImage(mmps_list)
```
